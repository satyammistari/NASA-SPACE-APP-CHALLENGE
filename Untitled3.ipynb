```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime 

# Initialize the IRIS client
client = Client("IRIS")

# Define the time window for the seismic event
start_time = UTCDateTime("2018-11-30T19:05:46.039000Z")
end_time = UTCDateTime("2018-11-30T19:06:31.039000Z")  # 45 seconds after start time

# Specify the station, location, and channel
network = "XB"   # Example Network identifier
station = "ELYH0" # Station identifier (Mars or simulation location)
location = "02"  # Location code from the SID "XB_ELYH0_02_VMA"
channel = "VMA"  # Channel identifier

# Download the waveform data
st = client.get_waveforms(network, station, location, channel, start_time, end_time)

# Save the data to a MiniSEED file
st.write("seismic_event.mseed", format="MSEED")

# Plot the waveform
st.plot()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import detrend, butter, filtfilt

# Create synthetic seismic data and save it to a CSV file
data = {
    'Time': pd.date_range(start='2018-11-30T19:05:46.039000Z', periods=75, freq='4S'),
    'Sample': np.random.randint(38016, 38024, size=75) + np.random.normal(0, 1, 75)  # Random noise added
}
df = pd.DataFrame(data)
df.to_csv('synthetic_seismic_data.csv', index=False)

# Load seismic data from the synthetic CSV file
file_path = 'synthetic_seismic_data.csv'
data = pd.read_csv(file_path)

# Convert the 'Time' column to datetime format
data['Time'] = pd.to_datetime(data['Time'])

# Check for and handle missing values
data.fillna(method='ffill', inplace=True)  # Forward fill to handle NaN values

# Detrend the sample data
data['Sample'] = detrend(data['Sample'].values)

# Define a Butterworth filter
def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = filtfilt(b, a, data)
    return y

# Filter the sample data
fs = 1 / 4  # Sample rate in Hz (4 seconds)
cutoff = 0.1  # Desired cutoff frequency of the filter in Hz
data['Filtered_Sample'] = lowpass_filter(data['Sample'], cutoff, fs)

# Plot the original and filtered signals
plt.figure(figsize=(12, 6))
plt.plot(data['Time'], data['Sample'], label='Original Signal', alpha=0.5)
plt.plot(data['Time'], data['Filtered_Sample'], label='Filtered Signal', color='red')
plt.title('Seismic Wave Data - Original and Filtered Signals')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()  # Improve layout
plt.show()
```
